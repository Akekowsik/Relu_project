```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
import time
import io
 
df = pd.read_csv('/content/data.csv')
asin = df['Asin']
cn = df['country']
lis = []
st = time.time()
for i in range(100):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'}
    html_text=requests.post(f'https://www.amazon.{cn[i]}/dp/{asin[i]}', headers=headers)
    if html_text.status_code == 200:
        soup = BeautifulSoup(html_text.text,'lxml')
        data = {}
        Product_Title=soup.find(id='productTitle').text.strip()
        Product_Image_URL=soup.find_all('img')[0]['src']
        Price_of_the_Product=soup.find_all('a', href="javascript:void(0)")[1].text.strip()

        Product_Details=soup.find_all('span',class_='a-list-item')
        description = ''
        for i in Product_Details:
          if i:
            description += i.text.replace('/n','') + '\n'
        description.replace(' ','')
        data["pt"]=Product_Title
        data["piu"]=Product_Image_URL
        data["pp"]=Price_of_the_Product
        data["pd"]=description
        lis.append(data)
    else:
        lis.append(f'https://www.amazon.{cn[i]}/dp/{asin[i]} not available')
for i in range(100):
    print(lis[i])
    print('\n')
et = time.time()
print(et-st)
```
